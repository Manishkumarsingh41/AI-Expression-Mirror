```python
# ✅ Step 1: Set paths (change if your file names differ)
source_image_path = "/content/drive/MyDrive/face-animation/image.jpg"
driving_video_path = "/content/drive/MyDrive/face-animation/drive.mp4"
checkpoint_path = "/content/drive/MyDrive/face-animation/first-order-model/vox-cpk.pth.tar"
config_path = "/content/drive/MyDrive/face-animation/first-order-model/config/vox-256.yaml"
repo_path = "/content/drive/MyDrive/face-animation/first-order-model"

# ✅ Step 2: Import libraries
import sys
sys.path.append(repo_path)

from demo import load_checkpoints, make_animation
from skimage import img_as_ubyte
import imageio
import numpy as np
from skimage.transform import resize
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")

from base64 import b64encode
from IPython.display import HTML

# ✅ Step 3: Load image and driving video
source_image = imageio.imread(source_image_path)
reader = imageio.get_reader(driving_video_path)
fps = reader.get_meta_data()['fps']

driving_video = []
for frame in reader:
    driving_video.append(resize(frame, (256, 256))[..., :3])
reader.close()

source_image = resize(source_image, (256, 256))[..., :3]

# ✅ Step 4: Display source image and first frame of driving video
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(source_image)
axs[0].set_title('Source Image')
axs[1].imshow(driving_video[0])
axs[1].set_title('First Frame of Driving Video')
for ax in axs:
    ax.axis('off')
plt.tight_layout()
plt.show()

# ✅ Step 5: Load model
generator, kp_detector = load_checkpoints(config_path=config_path,
                                          checkpoint_path=checkpoint_path)

# ✅ Step 6: Generate animation
predictions = make_animation(source_image, driving_video, generator, kp_detector, relative=True)

# ✅ Step 7: Save output
output_path = "/content/drive/MyDrive/face-animation/result.mp4"
imageio.mimsave(output_path, [img_as_ubyte(frame) for frame in predictions], fps=fps)

# ✅ Step 8: Display result in notebook
mp4 = open(output_path, 'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML(f'<video width=400 controls><source src="{data_url}" type="video/mp4"></video>')

print("✅ Done! Animation saved at:", output_path)
```
